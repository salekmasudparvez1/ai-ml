````python
import numpy as np
import pandas as pd

np.random.seed(24)

num_employees = 50  # you can increase this if needed

# Employee data
emp_ids = np.arange(101, 101 + num_employees)
emp_ages = np.random.randint(20, 55, num_employees)
emp_income = np.random.randint(35000, 95000, num_employees)

# Fixed set of roles (repetition allowed)
roles_list = [
    "HR Manager",
    "Software Developer",
    "Accountant",
    "Marketing Executive",
    "Sales Executive",
    "Backend Developer",
    "HR Executive",
    "Financial Analyst",
    "Digital Marketer",
    "Operations Manager"
]

emp_roles = np.random.choice(roles_list, num_employees)

# Create DataFrame
data = pd.DataFrame({
    "emp_id": emp_ids,
    "emp_age": emp_ages,
    "emp_salary": emp_income,
    "role": emp_roles
})

# Save dataset
data.to_csv("staff_data.csv", index=False)
````

# Q4 - Create a new NumPy array where each employee’s age is adjusted based on:
- Add 4 years if age is less than the average age
- Add 2 years if age is greater than or equal to the average age
Do NOT modify the original array.
### Print:
```bash
original array
updated array
average age used for condition
```


````python
old_data = np.array(data)
new_numpy_arr = old_data.copy()
old_avg_age = old_data[:,1].mean()

new_age =np.where(
    new_numpy_arr[:,1] < old_avg_age,
    new_numpy_arr[:,1]+4,
    new_numpy_arr[:,1]+2
)
new_numpy_arr[:,1] = new_age

print("original array",old_data)
print("updated array",new_numpy_arr)
print("average age used for condition",old_avg_age)
````

# Q5-Using NumPy:
- Assume all salaries of Tech roles increase by 8%
- Calculate the new total salary expense
- Find the overall increase in expense after the raise
---
### ***Print both old and new totals along with the difference.***








````python
employ_data = np.array(data)
old_expense =np.sum(employ_data[:,2])
update_salary = np.where(
    (employ_data[:,3]=='Software Developer')|(employ_data[:,3]=='Backend Developer')|((employ_data[:,3]=='Digital Marketer')),
    employ_data[:,2]+(employ_data[:,2]*0.08),
    employ_data[:,2]
)
employ_data[:,2]=update_salary
new_expense =np.sum(employ_data[:,2])
print('Old expense total =',old_expense)
print('New expense total =',new_expense)
print('expense increase =',new_expense-old_expense)
````

# **Q10 - Using .loc:**
- Increase salary by 12% for "Software Developer" and "Backend Developer"
- Increase salary by 5% for all other roles
### Then:
- Print the dataframe that includes role , previous_salary , new_salary , increase_of_salary


````python
import pandas as pd
old_df = pd.DataFrame(data)
new_df = old_df.copy()
new_df["emp_salary"] = new_df["emp_salary"].astype(float)
previous_salary = old_df['emp_salary']
find_increase_salary_employ = new_df['role'].isin([
    "Software Developer",
    "Backend Developer"
    ])
find_decrease_salary_employ = ~find_increase_salary_employ
# find_decrease_salary_employ = new_df['role'].isin([
#     "HR Manager",
#     "Accountant",
#     "Marketing Executive",
#     "Sales Executive",
#     "HR Executive",
#     "Financial Analyst",
#     "Digital Marketer",
#     "Operations Manager"
#     ])
new_df.loc[find_increase_salary_employ,"emp_salary"]=new_df.loc[find_increase_salary_employ,"emp_salary"]+new_df.loc[find_increase_salary_employ,"emp_salary"]*0.12

new_df.loc[find_decrease_salary_employ,"emp_salary"]=new_df.loc[find_decrease_salary_employ,"emp_salary"]+new_df.loc[find_decrease_salary_employ,"emp_salary"]*0.05

increase_of_salary = new_df['emp_salary']-previous_salary
new_salary = new_df["emp_salary"]


final=pd.DataFrame({
    "role":new_df["role"],
    "previous_salary":previous_salary,
    "new_salary":new_salary,
    "increase_of_salary":increase_of_salary


})
final
````
