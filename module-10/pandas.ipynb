## 1️⃣ dropna() — Handle Missing Data

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'name': ['A', 'B', 'C', 'D', 'E'],
    'salary': [100, None, 200, None, 300]
})

print('Drop NaN from column:')
print(df['salary'].dropna())

print('\nDrop rows with ANY NaN:')
print(df.dropna())

print('\nDrop rows where ALL NaN:')
print(df.dropna(how='all'))

print('\nDrop based on salary column:')
print(df.dropna(subset=['salary']))

print('\nFill NaN:')
print(df['salary'].fillna(0))
```

## 2️⃣ duplicated() — Advanced Duplicate Detection

```python
df = pd.DataFrame({
    'employee': ['A', 'B', 'C', 'D', 'E', 'F'],
    'salary': [100, 200, 200, 300, 300, 300]
})

print(df)

print('\nDuplicate flags (default keep=first):')
print(df['salary'].duplicated())

print('\nKeep last:')
print(df['salary'].duplicated(keep='last'))

print('\nAll duplicates:')
print(df[df['salary'].duplicated(keep=False)])

print('\nFull row duplicates:')
print(df[df.duplicated(keep=False)])

# Multi-column example
df2 = pd.DataFrame({
    'name': ['A','A','B','B'],
    'salary': [100,100,200,300]
})

print('\nMulti-column duplicates:')
print(df2.duplicated(subset=['name','salary']))
```

## 3️⃣ value_counts() — Frequency Analysis

```python
df = pd.DataFrame({
    'salary': [100, 200, 200, 300, 300, 300, 400]
})

counts = df['salary'].value_counts()

print('Frequency:')
print(counts)

print('\n>=2 occurrences:')
print(counts[counts >= 2])

print('\nExactly 3 times:')
print(counts[counts == 3])

print('\n>3 times:')
print(counts[counts > 3])

print('\nTop 2 most frequent:')
print(counts.head(2))

print('\nLeast frequent:')
print(counts.tail(1))

print('\nMode (safe):')
print(df['salary'].mode())

print('\nidxmax (fast):')
print(counts.idxmax())

print('\nPercentage:')
print(df['salary'].value_counts(normalize=True))

# get rows where value appears exactly 3 times
print('\nRows with salary appearing 3 times:')
print(df[df['salary'].map(counts) == 3])

# one-liner version
print('\nOne-liner:')
print(df[df['salary'].map(df['salary'].value_counts()) == 3])

# include NaN in counts
df2 = pd.DataFrame({'salary':[100, None, 100]})
print('\nInclude NaN:')
print(df2['salary'].value_counts(dropna=False))
```

## 4️⃣ shift() — Consecutive Pattern Detection

```python
df = pd.DataFrame({
    'num': [1, 1, 1, 2, 2, 3, 3, 3, 3]
})

df['prev1'] = df['num'].shift(1)
df['prev2'] = df['num'].shift(2)

print(df)

print('\n3 consecutive numbers:')
result = df[(df['num'] == df['prev1']) & (df['num'] == df['prev2'])]['num'].drop_duplicates()
print(result)

# General N consecutive
n = 3
mask = True
for i in range(1, n):
    mask &= df['num'] == df['num'].shift(i)

print('\nGeneral N consecutive:')
print(df[mask]['num'].drop_duplicates())
```

## 🔥 BONUS: GroupBy (SQL Style)

```python
df = pd.DataFrame({
    'salary': [100, 200, 200, 300, 300, 300, 400]
})

grouped = df.groupby('salary').size()

print(grouped[grouped == 3])
```

## 🔥 BONUS: Multi-column Frequency

```python
df = pd.DataFrame({
    'name': ['A','A','B','B','B'],
    'city': ['X','X','Y','Y','Z']
})

print(df.value_counts(['name','city']))
```
