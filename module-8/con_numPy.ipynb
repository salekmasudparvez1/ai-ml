## 1. The Conversion Workflow
In real-world data science, data rarely arrives as a NumPy array. It often comes as JSON (Dictionaries), unique IDs (Sets), or immutable configurations (Tuples). 

### Conversion Matrix
| Source Type | To NumPy | Note |
| :--- | :--- | :--- |
| **List** | `np.array(list)` | Most common, maintains order. |
| **Tuple** | `np.array(tuple)` | Direct conversion, useful for fixed constants. |
| **Set** | `np.array(list(set))` | **Warning:** Sets are unordered; order is not guaranteed. |
| **Dict** | `np.array(list(d.keys()))` | Requires selecting keys, values, or items. |
| **Dict** | `np.array(list(d.values()))` | Requires selecting keys, values, or items. |
| **Dict** | `np.array(list(d.items()))` | Requires selecting keys, values, or items. |

---

```python
import numpy as np

# 1. Unordered to Ordered (Set -> NumPy)
# Note: We must cast to a list first because sets don't support indexing/sequencing directly in the same way
s = {4, 1, 3, 2, 2} # Sets remove duplicates automatically
np_from_set = np.array(list(s))

print(f"Original Set: {s}")
print(f"NumPy Array:  {np_from_set} | Dtype: {np_from_set.dtype}")
```

## 2. Sequence Conversions (Lists & Tuples)
Lists and Tuples are the most straightforward to convert. 

### 💡 Pro Tip: `np.asarray()` vs `np.array()`
* Use `np.array()` to create a **copy** of the data.
* Use `np.asarray()` if you want to save memory; it won't copy the data if the input is already an ndarray.

```python
lst = [10.5, 20.1, 30.7]
tpl = (1, 2, 3)

arr_list = np.array(lst)   # Floats
arr_tuple = np.array(tpl)  # Integers

print("From List:", arr_list)
print("From Tuple:", arr_tuple)
```

## 3. Mapping Conversions (Dictionaries)
Dictionaries are key-value pairs. To convert them, you must decide which part of the data you need.

| Method | Target |
| :--- | :--- |
| `.keys()` | Features / Column Names |
| `.values()` | Data Observations |
| `.items()` | Key-Value pairs (creates a 2D array) |

```python
data_map = {'Jan': 150, 'Feb': 200, 'Mar': 180}

# Extracting values as a clean numeric vector
values_arr = np.array(list(data_map.values()))

# Extracting as pairs (Note: NumPy will often cast these to strings if types are mixed)
pairs_arr = np.array(list(data_map.items()))

print("Values Only:", values_arr)
print("\nPairs Array (2D):\n", pairs_arr)
```

## 4. Summary Guide

1. **Heterogeneous Data:** Python lists can hold multiple types (`[1, "a"]`). NumPy will **upcast** everything to a string if you do this. Avoid mixed types.
2. **Memory:** Large lists converted to NumPy save significant RAM because NumPy stores data in contiguous memory blocks.
3. **Iteration:** Once in NumPy, stop using `for` loops. Use **vectorization** for 100x speedups.
