```python
# 1. Set to List / Array
import numpy as np
s = {1,2,3,4}
print('Set:', s)
print('List:', list(s))
print('NumPy:', np.array(list(s)))
```

```python
# 2. List to Tuple / Set / NumPy
lst = [1,2,2,3]
print('List:', lst)
print('Tuple:', tuple(lst))
print('Set:', set(lst))
print('NumPy:', np.array(lst))
```

```python
# 3. Tuple to List / Set / NumPy
t = (10,20,30)
print('Tuple:', t)
print('List:', list(t))
print('Set:', set(t))
print('NumPy:', np.array(t))
```

```python
# 4. Dictionary to Keys / Values / Items / NumPy
d = {'a':10,'b':20,'c':30}
print('Dict:', d)
print('Keys:', list(d.keys()))
print('Values:', list(d.values()))
print('Items:', list(d.items()))
print('NumPy Values:', np.array(list(d.values())))
print('NumPy Items:', np.array(list(d.items())))
```

```python
# 5. Summary All to NumPy Array
print(np.array([1,2,3]))
print(np.array((1,2,3)))
print(np.array(list({1,2,3})))
print(np.array(list({'x':1,'y':2}.values())))
```
