## 1. Core Allocation Functions
NumPy provides three primary ways to allocate raw memory for arrays. The choice depends on whether you need initialized or uninitialized data.

| Function | Initial Value | Performance | Use Case |
| :--- | :--- | :--- | :--- |
| `np.empty()` | **Garbage** (Random) | 🚀 Fastest | When you will overwrite every element immediately |
| `np.zeros()` | **0.0** | 🟢 Balanced | Default choice for counters or accumulators |
| `np.ones()` | **1.0** | 🟡 Slower | Multiplicative masks or weight initialization |

### 💡 Pro Tip: The Speed of `empty` 
Unlike `zeros` or `ones`, `np.empty` does **not** set the memory values. It simply claims a block of RAM. If that RAM previously held data from another application, you will see those "garbage" values.

````python
import numpy as np

# Fast allocation (3x3 matrix)
raw_buffer = np.empty((3, 3))
print("Empty (Memory Garbage):\n", raw_buffer)

# Safe allocation with specific type
zero_fill = np.zeros((3, 3), dtype=np.int32)
print("\nZeros (Cleaned):\n", zero_fill)
````

## 2. Context-Aware Creation (`_like` family)
When working with existing pipelines, you often need a new array that perfectly matches the **shape** and **dtype** of an input. 

Instead of manually passing `shape=arr.shape`, use the `_like` variants. This is cleaner, faster, and more readable.

* **`np.empty_like(x)`**: Matches shape/type of `x`, contains garbage.
* **`np.zeros_like(x)`**: Matches shape/type of `x`, filled with 0.
* **`np.ones_like(x)`**: Matches shape/type of `x`, filled with 1.

---

````python
# Create a template array
template = np.array([[10, 20], [30, 40]], dtype=np.float32)

# The 'Pro' way to create a mask for this array
mask = np.ones_like(template)

print(f"Template Shape: {template.shape} | Dtype: {template.dtype}")
print("Generated Mask:\n", mask)
````

## 3. Understanding Dimensions
In NumPy, dimensions follow a specific hierarchy. Understanding this is key to high-level data science:

1.  **1D:** `(n,)` — A single vector.
2.  **2D:** `(rows, cols)` — A standard matrix.
3.  **3D:** `(depth, rows, cols)` — Think of "depth" as pages or layers (e.g., RGB channels in an image).

> **Mathematical Notation:**
> An array $A$ with shape $(m, n)$ belongs to the space $\mathbb{R}^{m \times n}$.

## 4. Performance & Memory Layout (`order`)
For high-performance computing, you can define how data sits in your RAM using the `order` parameter.

* **`order='C'`**: (Default) C-style row-major storage. Fast for row-wise operations.
* **`order='F'`**: Fortran-style column-major storage. Better for column-wise operations.

```python
# Forcing Fortran order for specific linear algebra math
f_array = np.zeros((100, 100), order='F')
```

## 5. Summary Checklist

* **Beginner Rule:** Use `np.zeros()` to avoid bugs from random memory values.
* **Intermediate Rule:** Use `_like` functions to keep code DRY (Don't Repeat Yourself).
* **Advanced Rule:** Use `np.empty()` for massive pre-allocations where every microsecond counts.
