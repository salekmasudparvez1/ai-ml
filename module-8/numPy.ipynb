## 🚀 1. NumPy Introduction

```python
import numpy as np

numpy_array = np.array([1,2,3,4,4])
print(type(numpy_array))
```

## 🚀 2. NumPy vs Python List (Important Concept)

```python
num_list = [1, 2, 3, 4]
num_array = np.array([1, 2, 3, 4])

print("List * 5 =>", num_list * 5)
print("Array * 5 =>", num_array * 5)
```

## 🧠 3. NumPy Type Upcasting (VERY IMPORTANT)

```python
arr_1d = np.array([1, 2, 3, 4, 5])
print("int array:", arr_1d.dtype)

# 🔼 Upcasting: int -> float
arr_1d1 = np.array([1, 2, 3, 4, 5.9])
print("int + float =>", arr_1d1.dtype)

# 🔼 Upcasting: number -> string
arr_1d2 = np.array([1, 2, 3, 4, "5"])
print("int + string =>", arr_1d2.dtype)

# 🔼 Manual dtype control
arr_1d3 = np.array([1, 2, 3, 4, 5], dtype=np.int8)
print("forced int8 =>", arr_1d3.dtype)
```

## 📊 4. Array Properties (Shape, Dim, Type)

```python
arr = np.array([1,2,3,4,5])
print(arr)
print("shape:", arr.shape)
print("dim:", arr.ndim)
print("dtype:", arr.dtype)
```

## 📐 5. 1D, 2D, 3D Arrays

```python
# 1D
a1 = np.array([1,2,3,4,5])

# 2D
a2 = np.array([[1,2,3],[4,5,6]])

# 3D
a3 = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])

print(a1.shape, a1.ndim)
print(a2.shape, a2.ndim)
print(a3.shape, a3.ndim)
```

## 🔥 6. Special Cases in NumPy

```python
# Scalar (0D)
scalar = np.array(42)

# Boolean array
bool_arr = np.array([True, False, True])

# Empty array
empty = np.array([])

# NaN and INF
special = np.array([1, np.nan, np.inf, -np.inf])

print(scalar, scalar.ndim)
print(bool_arr)
print(empty)
print(special)
```

## 🔁 7. Reshape, Identity, Transpose

```python
arr = np.array([1,2,3,4,5,6])
print(arr.reshape(2,3))

m = np.array([[1,2,3],[4,5,6]])
print("transpose:\n", m.T)

print("identity:\n", np.eye(3))
```
