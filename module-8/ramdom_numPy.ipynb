```python
import numpy as np
from numpy import random
import string
from collections import Counter
import math
```

# 🎲 PART 1: RANDOM NUMBERS IN NUMPY

```python
print('random float:', random.random())
print('chance event:', random.random() < 0.5)
print('dice:', random.randint(1,6))
print('otp:', random.rand(100000,999999))
```

```python
print('uniform [0, 1):', random.uniform())
print('uniform [10, 20):', random.uniform(10, 20))
print('range integers:', random.randint(1, 11, size=5))
print('dice roll batch:', random.randint(1, 7, size=10))
```

# 🧱 PART 2: NDARRAY BASICS (SHAPE • DTYPE • SIZE)

```python
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([[1,2,3],[4,5,6]])
arr3 = np.array([[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]])

print("arr1.ndim:", arr1.ndim)
print("arr2.ndim:", arr2.ndim)
print("arr3.ndim:", arr3.ndim)

print("arr1.shape:", arr1.shape)
print("arr2.shape:", arr2.shape)
print("arr3.shape:", arr3.shape)

print("arr1.dtype:", arr1.dtype)
print("arr2.dtype:", arr2.dtype)
print("arr3.dtype:", arr3.dtype)

print("arr1.size:", arr1.size)
print("arr2.size:", arr2.size)
print("arr3.size:", arr3.size)
```

# 🧩 PART 3: COPY VS REFERENCE

```python
base_arr = np.array([10, 20, 30, 40])
reference_arr = base_arr
copied_arr = base_arr.copy()

reference_arr[0] = 999

print('original array:', base_arr)
print('copied array   :', copied_arr)
print('same object?   :', base_arr is reference_arr)
print('copy object?   :', base_arr is copied_arr)
```

# 🔥 PART 4: ARRAY CREATION FUNCTIONS

```python
print(np.zeros((2,3)))
print(np.ones((3,3)))
print(np.full((2,3),7))
print(np.empty((2,2)))
```

# 🔄 PART 5: DATA TYPE SYSTEM

```python
print(np.array([1,2,3]).dtype)
print(np.array([1.5,2.4,3.2]).dtype)
print(np.array([1,2,3.2,'hello']).dtype)
```

# 📊 PART 6: ADVANCED INDEXING & ITERATION

```python
advanced = np.array([[10,20,30],[40,50,60],[70,80,90]])

print('selected diagonal-like values:', advanced[[0,1,2],[0,1,2]])
print('picked rows:', advanced[[0,2]])
print('values above 45:', advanced[advanced > 45])

print('iterating row by row:')
for row in advanced:
    print(row)

print('iterating element by element:')
for v in advanced.flat:
    print(v, end=' ')
print()
```
