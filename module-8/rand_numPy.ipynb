## 1. Random Distribution Comparison
NumPy offers several ways to generate random data. Choosing the right one depends on your mathematical requirements.

| Function | Output Type | Range | Arguments |
| :--- | :--- | :--- | :--- |
| `np.random.rand()` | Float | $[0, 1)$ | `d0, d1, ...` (Individual dims) |
| `np.random.uniform()` | Float | $[low, high)$ | `low, high, size` |
| `np.random.randint()` | Integer | $[low, high)$ | `low, high, size` |

---

### `np.random.rand` vs `np.random.uniform` 
`rand` is a shortcut for a uniform distribution between 0 and 1. `uniform` allows you to define your own boundaries (e.g., between 10 and 50).

```python
import numpy as np

# 1. Basic Rand (0 to 1)
print("Rand (2x2):\n", np.random.rand(2, 2))

# 2. Uniform (Custom Range: 5 to 10)
print("\nUniform (5 to 10):\n", np.random.uniform(5, 10, size=(2, 2)))

# 3. Randint (Whole numbers: 0 to 99)
print("\nRandint (0-99):\n", np.random.randint(0, 100, size=5))
```

## 2. Linear Spacing (`np.linspace`)
Unlike the random functions, `linspace` (Linear Space) creates a **predictable**, evenly spaced sequence of numbers.

**Key Difference:** 
* `np.arange(start, stop, step)`: You define the **distance** between numbers.
* `np.linspace(start, stop, num)`: You define **how many** numbers you want in total.



**Note:** By default, `linspace` **includes** the stop value (endpoint=True).

```python
# Create 10 evenly spaced points between 0 and 100
linear_data = np.linspace(0, 100, 10)

print("Linspace Output:", linear_data)
print("Distance between each point is identical.")
```

## 3. Practical Use Case: Visualizing Data
In professional workflows, we use `linspace` to create an X-axis and `random` or `uniform` to simulate noise or data points on the Y-axis.

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 50)
y = np.random.uniform(0, 5, 50) # Simulated random noise
# x1 = np.array([1,2,3,4])
# y1 = np.array([3,7,8,9])
plt.scatter(x, y)
# plt.scatter(x1, y1)
plt.title("Linspace (X) vs Uniform Random (Y)")
plt.show()
```

## 4. Summary Guide

1. **Need Whole Numbers?** Use `randint`.
2. **Need Random Decimals?** Use `uniform` (or `rand` for 0-1).
3. **Need a Graph Axis?** Use `linspace`.
4. **Pro Tip:** In modern NumPy (v1.17+), it is recommended to use the new Generator API: `rng = np.random.default_rng()` for better statistical performance.
